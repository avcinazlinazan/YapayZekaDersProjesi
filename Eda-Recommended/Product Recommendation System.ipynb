```python
import pandas as pd
import numpy as np
import neattext.functions as nfx
import seaborn as sn

from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity,linear_kernel
```

```python
df = pd.read_csv('urunTavsiye.csv')
df.head()
```

```python
#df.drop("Unnamed: 0.1",axis=1,inplace=True)
```

```python
df.head()
```

```python
# list all the methods present in the neattext function

dir(nfx)
```

```python
df['product_Info'].iloc[1:5]
```

```python
# generating clean text by removing the stopwords and special characters


df['product_Info'] = df['productInfo'].apply(nfx.remove_stopwords)

df['product_Info'] = df['product_Info'].apply(nfx.remove_special_characters)

df['product_Info'].iloc[1:5]
```

## Vectorise the clean productInfo

```python
# vectorizing the product_Info

countvect = CountVectorizer()

cv_mat = countvect.fit_transform(df['product_Info'])

cv_mat
```

```python
cv_mat.todense()
```

```python
df_cv_words = pd.DataFrame(cv_mat.todense(),columns=countvect.get_feature_names_out())

df_cv_words
```

##  cosine similarity 

```python
# cosine similarity matrix

cosine_sim_mat = cosine_similarity(cv_mat)

cosine_sim_mat
```

```python
# drop duplicates 

product_index = pd.Series(df.index,index = df['product_Info']).drop_duplicates()
```

```python
def recommend_product(product_Info,numrec = 10):
    product_index = pd.Series(
        df.index, index=df['product_Info']).drop_duplicates()
    index = product_index[product_Info]
    scores = list(enumerate(cosine_sim_mat [index]))

    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)

    selected_product_index = [i[0] for i in sorted_scores[1:]]

    selected_product_score = [i[1] for i in sorted_scores[1:]]

    rec_df = df.iloc[selected_product_index]

    rec_df['Similarity_Score'] = selected_product_score

    final_recommended_product = rec_df[[
        'product_Info', 'Similarity_Score', 'productLink', 'productPrice','productName']]

    return final_recommended_product.head(numrec)

ans = recommend_product('Thinkbook 16 21cy004rtx I71260p 
                        16gb 512ssd Rtx2050 16 Qhd Freedos 
                        Tanabilir Bilgisayar',20)
ans
    
```
